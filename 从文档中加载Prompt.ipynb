举例1：从物理磁盘的yaml文件中读取prompt数据

```python
from langchain_core.prompts import load_prompt
import dotenv

dotenv.load_dotenv()

prompt = load_prompt("prompt.yaml", encoding="utf-8")
# print(prompt)
print(prompt.format(var1="1", var2="2"))
```
