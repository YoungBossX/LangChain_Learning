# 2、角度1：非对话模型的调用

```python
import os
import dotenv
from langchain_openai import OpenAI
dotenv.load_dotenv()
os.environ["LLM_API_KEY"] = os.getenv("LLM_API_KEY")
os.environ["LLM_BASE_URL"] = os.getenv("LLM_BASE_URL")
```

```python
llm = OpenAI(model=os.getenv("LLM_MODEL_ID"))
str=llm.invoke("介绍一下你自己。")
print(str)
print(type(str))
```
